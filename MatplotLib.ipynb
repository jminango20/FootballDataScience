## Elements of Matplotlib –What’s What?

```
import matplotlib.pyplot as plt
import numpy as np

```

```
#Creo el grafico
fig, ax = plt.subplots(figsize=(8,8))
fig.set_facecolor("white")
ax.grid(linewidth=0.25, zorder=1)
ax.set_title("Este es un titulo usando Matplotlib")

#Genero los datos
np.random.seed(402)
x = np.random.uniform(0,1,50)
y = np.random.uniform(0,1,50)

ax.scatter(x,y,
           edgecolors="white",
           linewidth=0.5,
           s=400,
           zorder=2)

#Limitar mis ejes
ax.set_xlim(0,1)
ax.set_ylim(0,1)

plt.tight_layout() #Ajustar mi grafico
plt.savefig("../3_default_mpl", bbox_inches="tight",dpi=300)
plt.show()
```

## Elements of Matplotlib – Figures and Axes

```
#Other Example
fig, ax = plt.subplots(figsize=(8,8))

fig.set_facecolor("#444444")
ax.patch.set_facecolor("grey")

fig.text(0.05,1,"This is the figure",fontsize=20)
ax.text(0.5,0.5,"This is the text object",ha="center",fontsize=20)

#Atencions
plt.tight_layout()
plt.show()
```

## Lines

```
# Primer caso: We can plot together
x = [0,0.1,0.2,0.3,0.4]
y = [0.2,0.8,0.7,0.9,1]

plt.plot(x,y,color="red")
```

```
# Segundo caso: If we want to plot individual lines with starting and ending x and y values

#using x and y from above. Por ejemplo para pases
ex = [0.9,0.4,0.3,0.7,0.1]
ey = [0.1,0.3,0.6,0.1,0.9]

for x,y,ex,ey in zip(x,y,ex,ey):
  plt.plot([x,ex],[y,ey],color="blue")
```

## Ticks and Spines

```
fig, ax = plt.subplots(figsize=(8,8))
fig.set_facecolor("white")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#Change color axis
ax.tick_params(axis="x",color="blue",length=10,width=2) #ticks - lineas
ax.tick_params(axis="y",length=10,labelcolor="blue",labelsize=16,labelrotation=45)
```

* We can adjust these by using `mpl.rcParams[<param_name>] = value`
* We can revert to default matplotlib by using `mpl.rcParams.update(mpl.rcParamsDefault)`

The main params we are interested in are:
* axes.facecolor
* axes.labelcolor
* figure.facecolor
* font.family
* xtick.color

## Fonts

```
import matplotlib.font_manager
from IPython.core.display import HTML
```

```
def make_html(fontname, typetest_string):
  return "<p><b>{font}:</b> <br/><span style='font-family:{font}; font-size: 20px;'>{typetest_string}</p><br>".format(typetest_string=typetest_string, font=fontname)
```

```
typetest_string = "The quick brown fox jumps over the lazy dog"
```

```
code = "\n".join([make_html(font,typetest_string) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])
```

```
HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

## Changing fonts

```
title_font = "Alegreya Sans"
main_font = "Open Sans"
```

```
np.random.seed(402)
x = np.random.uniform(0,1,100)
y = np.random.uniform(0,1,100)

```

```
fig, ax = plt.subplots(figsize=(8,8))
fig.set_facecolor("w")
ax.patch.set_color("w")

ax.set_title("This is a title using default matplolib",loc='left',fontweight="bold",fontsize=18,fontfamily=title_font)
ax.grid(lw=0.5, color='k', zorder=1)

ax.scatter(x,y,color="dodgerblue", edgecolors="w", linewidths=1, s=400, zorder=2)

ax.text(x[19],y[19]+0.025,"Point label",ha="center",color="k", fontsize=14, fontweight="medium", zorder=3, fontfamily=main_font)
ax.set_xlabel("This is an x label", fontweight="bold", fontfamily=title_font, size=16)
ax.set_ylabel("This is a y label", fontweight="bold", fontfamily=title_font, size=16)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
plt.show()
```

## The folly of not using zorder
