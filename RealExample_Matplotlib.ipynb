```
#Packages
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

#Company logos
from PIL import Image
import requests
from io import BytesIO


from highlight_text import htext
```

## Dataset

A dataset showing team goals for and
against per season in the Premier
League from 2010/11 to 2019/20

```
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/SoccerAnalysis/pl_goal_summary_data.csv')
df.head(10)
```

```
#Selecting Team
team = "Arsenal"
```

## The initial plot

```
#Creating Figure
fig, ax = plt.subplots(figsize=(8,8))
fig.set_facecolor("w")

#adding a grid
ax.grid()

#Getting our data to plot
x = df.goals_for.to_list()
y = df.goals_against.to_list()
#plotting
ax.scatter(x,y,s=120)

#Getting data for specified team
x = df[df.team==team].goals_for.to_list()
y = df[df.team==team].goals_against.to_list()
t = df[df.team==team].team_season.to_list()
#plotting a line to track team's change over season
ax.plot(x,y)

#Plotting team's most recent season
ax.scatter(x[-1],y[-1],s=120)
t = ax.text(x[-1],y[-1]-2,t[-1],color='k',fontsize=14,ha='center')

#adding title and subtitle
fig.text(0.05,1,"{}'s goal difference from 2010/11 to 2019/20\n".format(team), fontsize=16, fontweight="bold")
fig.text(0.05,1,"English Premier League", fontweight="regular", fontsize=14)

#adding xlabel and ylabels
ax.set_xlabel("Goals For", fontweight="bold", fontsize=14)
ax.set_ylabel("Goals Against", fontweight="bold", fontsize=14)

plt.tight_layout()
plt.show()
```

## Adding path effects

Note that we created our text label as a variable : `t`

`t = ax.text(x[-1], y[-1]-2, t[-1],
color=text_color,
fontsize=14, ha="center",
fontfamily=body_font)`

By setting that as a variable, we now have a Text object
that we can modify. By using path_effects we can draw a
border around the text to make it stand out from the
background:

`t.set_path_effects([path_effects.withStroke(linewidth=3,
foreground=background)])`

## An alternative: `rcParams`
Most of what we have covered has required us adding the same variables to
various parts of our code:
* `background = "#313332"`
* `text_color = "w"`
* `title_font = "Alegreya Sans"`

If we will reuse these elements, we can change the
`rcParams` to use those as our defaults. This saves us time and makes our output more consistent.

* `mpl.rcParams['figure.facecolor'] = "#313332"`
* `mpl.rcParams['axes.facecolor'] = "#313332"`
* `mpl.rcParams['font.family'] = "Alegrey Sans"`

Note that when we change `rcParams`, we change the default `matplotlib` until
either the package is reimported or reloaded, or until we reset our `rcParams`.

A simple way of doing this is to create a function to handle the reset:

`def reset_matplotlib():
  mpl.rcParams.update(mpl.rcParamsDefault)`
  
By creating this, we can simply reset rcParams to their default settings by running
`reset_matplotlib()`
